```python
import mlxtend
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
import numpy as np
import math
```

First dataset :

```python
dataset = [['Skirt', 'Sneakers', 'Scarf', 'Pants', 'Hat'],

    ['Sunglasses', 'Skirt', 'Sneakers', 'Pants', 'Hat'],

    ['Dress', 'Sandals', 'Scarf', 'Pants', 'Heels'],

    ['Dress', 'Necklace', 'Earrings', 'Scarf', 'Hat', 'Heels', 'Hat'],

   ['Earrings', 'Skirt', 'Skirt', 'Scarf', 'Shirt', 'Pants']]
```

```python
te=TransactionEncoder()
te_ary=te.fit(dataset).transform(dataset)    #Apply one-hot-encoding on our dataset
df=pd.DataFrame(te_ary, columns=te.columns_)  #Creating a new DataFrame from our Numpy array
df
```

```python
frequent_itemsets=apriori(df, min_support=0.6, use_colnames=True) #Letâ€™s select itemsets with a minimum of 60% support
frequent_itemsets
```

```python
association_rules(frequent_itemsets,metric="confidence",min_threshold=0.7) # associate itemsets with confidence over 70%.
```

```python
association_rules(frequent_itemsets,metric="lift",min_threshold=1.24)
```

We notice that skirt and pants are often bought together and should be sold together

Second data set :

```python
data=pd.read_csv('Market_Basket_Optimisation.csv', sep=',', header=None)
data.head()
```

```python
data.shape
```

```python
#Putting each customer's shopping items in a list

data2=[]
for i in range(0, 7501):
    data2.append([str(data.values[i,j]) for j in range(0, 20)])       
```

```python
#Eliminating nan from our list
data3=[]
temp=[]

for i in range(0, 7501):
    for j in range(0, 20):
        if(data2[i][j]!='nan'):
            temp.append(data2[i][j])
    data3.append(temp)
    temp=[]
```

```python
data3=np.array(data3,dtype=object)

```

```python
te=TransactionEncoder()
te_ary=te.fit(data3).transform(data3)    #Apply one-hot-encoding on our dataset
data=pd.DataFrame(te_ary, columns=te.columns_)  #Creating a new DataFrame from our Numpy array
data
```

```python
frequent_itemsets=apriori(data, min_support=0.05, use_colnames=True) #Letâ€™s select itemsets with a minimum of 5% support
frequent_itemsets
```

```python
association_rules(frequent_itemsets,metric="confidence",min_threshold=0.2) # associate itemsets with confidence over 20%.
```

```python
association_rules(frequent_itemsets,metric="lift",min_threshold=1.01)
```
